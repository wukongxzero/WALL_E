```python

def convert_cpp_to_puml(header_file_path):
    try:
        # Parse the header file
        cpp_header = CppHeaderParser.CppHeader(header_file_path)
    except Exception as e:
        return f"Error parsing: {e}"

    puml_lines = ["@startuml"]

    for class_name, class_obj in cpp_header.classes.items():
        # Handle Inheritance
        for base in class_obj["inherits"]:
            access = base.get("access", "public")
            puml_lines.append(f"{base['class']} <|-- {class_name}")

        puml_lines.append(f"class {class_name} {{")

        # Handle Properties (Variables)
        for access in ["private", "protected", "public"]:
            symbol = {"private": "-", "protected": "#", "public": "+"}[access]
            for prop in class_obj["properties"][access]:
                puml_lines.append(f"  {symbol} {prop['name']} : {prop['type']}")

        # Handle Methods (Functions)
        for access in ["private", "protected", "public"]:
            symbol = {"private": "-", "protected": "#", "public": "+"}[access]
            for method in class_obj["methods"][access]:
                # Skip the return type for constructors
                ret_type = method["rtnType"] if method["rtnType"] else ""
                params = ", ".join([f"{p['type']} {p['name']}" for p in method["parameters"]])
                puml_lines.append(f"  {symbol} {method['name']}({params}) : {ret_type}")

        puml_lines.append("}")

    puml_lines.append("@enduml")
    return "\n".join(puml_lines)

```

```python
import CppHeaderParser
import sys
import sys
import zlib
import base64
import string
from pathlib import Path

# PlantUML custom base64 alphabet mapping
PLANTUML_ALPHABET = string.ascii_uppercase + string.ascii_lowercase + string.digits + "-_"
STANDARD_ALPHABET = string.ascii_uppercase + string.ascii_lowercase + string.digits + "+/"
MAP_TABLE = str.maketrans(STANDARD_ALPHABET, PLANTUML_ALPHABET)

def encode_puml(puml_text):
    """Encodes PlantUML text into the format required for the web API."""
    zlibbed_str = zlib.compress(puml_text.encode('utf-8'))
    compressed_string = zlibbed_str[2:-4] # Remove zlib headers/checksums
    base64_string = base64.b64encode(compressed_string).decode('utf-8')
    return base64_string.translate(MAP_TABLE)

def convert_headers_to_puml(base_directory=".", pattern="*.h", output_filename="diagram.puml"):
    base_path = Path(base_directory)
    files = list(base_path.rglob(pattern))

def convert_headers_to_puml(base_directory=".", pattern="*.h", output_filename="diagram.puml"):
    # Initialize the base path
    base_path = Path(base_directory)
    
    # rglob(pattern) is shorthand for Path.glob(f"**/{pattern}")
    # It will find all files matching the pattern in the current dir and all subdirs
    files = list(base_path.rglob(pattern))
    
    if not files:
        print(f"No files found matching '{pattern}' in '{base_path.resolve()}'")
        return

    puml_lines = [
        "@startuml",
        f"' Generated from {base_path.resolve()}",
        "set separator none",
        "skinparam classAttributeIconSize 0" # Uses text (+/-) instead of icons
    ]

    for file_path in files:
        print(f"Processing: {file_path.relative_to(base_path)}")
        try:
            # CppHeaderParser requires a string path
            cpp_header = CppHeaderParser.CppHeader(str(file_path))
        except Exception as e:
            print(f"  [!] Skipped {file_path.name}: {e}")
            continue

        for class_name, class_obj in cpp_header.classes.items():
            # Handle Inheritance
            for base in class_obj.get("inherits", []):
                puml_lines.append(f"{base['class']} <|-- {class_name}")

            puml_lines.append(f"class {class_name} {{")

            # Handle Properties (Variables)
            for access in ["private", "protected", "public"]:
                symbol = {"private": "-", "protected": "#", "public": "+"}[access]
                for prop in class_obj["properties"][access]:
                    puml_lines.append(f"  {symbol} {prop['name']} : {prop['type']}")

            # Handle Methods (Functions)
            for access in ["private", "protected", "public"]:
                symbol = {"private": "-", "protected": "#", "public": "+"}[access]
                for method in class_obj["methods"][access]:
                    ret_type = method["rtnType"] if method["rtnType"] else ""
                    params = ", ".join([f"{p['type']} {p['name']}" for p in method["parameters"]])
                    puml_lines.append(f"  {symbol} {method['name']}({params}) : {ret_type}")

            puml_lines.append("}")

    puml_lines.append("@enduml")
    full_puml = "\n".join(puml_lines)

    # Save output
    output_path = Path(output_filename)
    output_path.write_text("\n".join(puml_lines))
    print(full_puml)

    # Generate Web Link
    encoded_url = encode_puml(full_puml)
    svg_url = f"http://www.plantuml.com/plantuml/svg/{encoded_url}"
    
    print(f"\nSuccess! Diagram saved to: {output_path.resolve()}")
    print(f"\nView/Download your SVG here:\n{svg_url}")
    
    print(f"\nSuccess! Processed {len(files)} files.")
    print(f"Diagram saved to: {output_path.resolve()}")
```
